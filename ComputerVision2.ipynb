```
import cv2
from google.colab.patches import cv2_imshow
```

```
img1 = cv2.imread('1.jpg',1)
img2 = cv2.imread('2.jpg',1)
```

```
img1.shape
```

```
img2.shape
```

```
# resizing the both images
image1=cv2.resize(img1,(500,300),interpolation = cv2.INTER_AREA)
image2=cv2.resize(img2,(500,300),interpolation = cv2.INTER_AREA)
```

```
# showing the image
cv2_imshow(image1)
cv2_imshow(image2)
```

```
# Difference btw the images
df=image1-image2
```

```
cv2_imshow(df)
```

```
df1=image2-image1
```

```
cv2_imshow(df1)
```

```
# Sum of two images
sum=image1+image2
```

```
cv2_imshow(sum)
```
