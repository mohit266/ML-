# **Reverse the image**

```
# Grab the dimensions of the image and calculate the center of the image
(h,w)=img.shape[:2]
center=(w/2,h/2)
```

```
# rotate the image by 180 degrees
rotation = cv2.getRotationMatrix2D(center,180,1)
rotated = cv2.warpAffine(img,rotation,(w, h))
cv2_imshow(rotated)
```
