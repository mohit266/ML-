```
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```
img1 = cv2.imread('1.jpg',1)
img2 = cv2.imread('2.jpg',1)
```

```
img1.shape
```

```
img2.shape
```

```
cv2_imshow(img1)
cv2_imshow(img2)
```

```
# resizing the image 1
image1 = cv2.resize(img1,(500,300), interpolation = cv2.INTER_AREA)
cv2_imshow(image1)
image1.shape
```

```
# resizing the image 2
image2 = cv2.resize(img2,(500,300),interpolation = cv2.INTER_AREA)
cv2_imshow(image2)  
image2.shape
```

```
# cropping the image1
crop_img1 = image1[50:300,100:500]
crop_img1.shape
```

```
# cropping the image 2
crop_img2 = image2[50:300,100:500]
crop_img2.shape
```

```
cv2_imshow(crop_img1)
cv2_imshow(crop_img2)
```

```
crop_img1[10:55,35:70]=img2[10:55,35:70]
crop_img2[10:55,35:70]=img1[10:55,35:70]
```

```
cv2_imshow(crop_img1)
cv2_imshow(crop_img2)
```
